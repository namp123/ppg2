```python
x=[]
anx2=anx.drop(['gender','age'],axis=1)
anx2=anx2.to_numpy()
anx2
```

```python
refl=np.array([2,1,1,1,1,1,1,1,1,2,1,1,1,1,1])
refl
```

```python
count=0
x=[]
for i in anx2:
    k=0
    count=0
    for j in i:
        while k<15:
            if j==refl[k]:
                count+=1
            k+=1
            break
    x.append(count)
```

```python
len(x)
```

```python
x=np.array(x)
x
```

```python
y=[] 
for i in x:
    if i>=7: #score of 7 or greater is classified as 1 that is HAS ANXIETY
        y.append(1)  
    else:    #score of lesser than 7 is classified as 0 that is DOES NOT HAVE ANXIETY
        y.append(0)
```

```python
y=np.array(y) #target array
y
```

```python
anx['score']=x
anx['target']=y
```

```python
anx
```

```python
anx=anx[anx['gender']!=3] #gender type 3:'other' not included for now
anx
```
